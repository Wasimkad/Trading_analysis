```python
def parse_string_to_list_dict(string_data: str):
    try:
        return ast.literal_eval(string_data)
    except (SyntaxError, ValueError) as e:
        print(f"Error parsing string: {e}")
        return None
```

```python
def get_new_df(df: pd.DataFrame) -> pd.DataFrame:
#to get the desired dataframe for easy implementation later
    new_df = []
    column_names = []
    count=0
    for row in df.itertuples(index=False):

        try:
            current_port = row.Port_IDs
            trade_history = row.Trade_History
            trade_history = parse_string_to_list_dict(trade_history)
    
            for trade in trade_history:
                current_row = []
                current_row.append(current_port)
    
                for key, value in trade.items():
                    current_row.append(value)
    
                column_names = ["Port_IDs"].extend(list(trade.keys()))
                new_df.append(current_row)
        except:
            count+=1
            pass

    print(count)#count helping us to get whether we skipped the nan 
    return pd.DataFrame(new_df, columns=column_names)
```

```python
def ROI(new_data: pd.DataFrame) -> pd.DataFrame:
    roi_results = []
    
    for port_id, group in new_data.groupby("Port_IDs"):
        # Total Investment (BUY trades)
        total_investment = group[group["side"] == "BUY"]["quantity"].sum()
        
        # Total Returns (SELL trades)
        total_returns = group[group["side"] == "SELL"]["quantity"].sum()
        
        # Total Fees (in USDT)
        total_fees = group["fee"].sum()
        
        # ROI Calculation
        if total_investment > 0:
            roi = ((total_returns - total_investment + total_fees) / (total_investment - total_fees)) * 100
        else:
            roi = 0
        
        roi_results.append({"Port_IDs": port_id, "ROI": roi})
    
    roi_df = pd.DataFrame(roi_results)
    
    return roi_df
```

```python
def PnL(new_data: pd.DataFrame) -> pd.DataFrame:
    pl_results = []
    
    for port_id, group in new_data.groupby("Port_IDs"):
        # Total Investment (BUY trades)
        total_investment = group[group["side"] == "BUY"]["quantity"].sum()
        
        # Total Returns (SELL trades)
        total_returns = group[group["side"] == "SELL"]["quantity"].sum()
        
        # Total Fees (in USDT)
        total_fees = group["fee"].sum()
        
        # Realized Profit
        realized_profit = group["realizedProfit"].sum()
        
        # Profit or Loss Calculation
        profit_or_loss = total_returns - total_investment + total_fees + realized_profit
        
        pl_results.append({"Port_IDs": port_id, "Profit/Loss": profit_or_loss})
    
    pl_df = pd.DataFrame(pl_results)
    
    return pl_df
```

```python
def sharpe_ratio(new_data: pd.DataFrame, risk_free_rate: float = 0.0, trading_days: int = 90) -> pd.DataFrame:

    sharpe_results = []
    
    # Created a copy and convert 'time' to datetime
    df = new_data.copy()
    df['datetime'] = pd.to_datetime(df['time'], unit='ms', errors='coerce')
    df['date'] = df['datetime'].dt.date
    
    # Processed each portfolio
    for port_id, group in df.groupby("Port_IDs"):
        # Prefered closed trades
        closed_trades = group[group["activeBuy"] == False]
        trades_to_use = closed_trades if not closed_trades.empty else group
        
        # Daily profit calculation
        daily_profit = trades_to_use.groupby("date")["realizedProfit"].sum().reset_index()
        
        # Fixed: Daily investment calculation
        buy_trades = trades_to_use[trades_to_use["side"] == "BUY"]
        # Changed the groupby operation to directly create a DataFrame with named columns
        daily_investment = (
            buy_trades.groupby("date")["quantity"]
            .sum()
            .to_frame(name="daily_investment")
            .reset_index()
        )
        
        # Merge daily profit and investment
        daily = pd.merge(daily_profit, daily_investment, on="date", how="left")
        daily["daily_investment"] = daily["daily_investment"].fillna(0)
        
        # Calculated daily returns
        daily["daily_return"] = np.where(
            daily["daily_investment"] > 0,
            daily["realizedProfit"] / daily["daily_investment"],
            0
        )
        
        # Calculated Sharpe ratio
        valid_returns = daily[daily["daily_investment"] > 0]["daily_return"]
        if len(valid_returns) < 2:
            sharpe = 0
        else:
            mean_return = valid_returns.mean()
            std_return = valid_returns.std()
            sharpe = ((mean_return - risk_free_rate) / std_return) * np.sqrt(trading_days) if std_return != 0 else 0
        
        sharpe_results.append({"Port_IDs": port_id, "Sharpe Ratio": sharpe})
    
    return pd.DataFrame(sharpe_results)
```

```python
def MDD(new_data: pd.DataFrame) -> pd.DataFrame:
    mdd_results = []
    
    for port_id, group in new_data.groupby("Port_IDs"):
        # Used only closed trades (activeBuy == False)
        closed_trades = group[group["activeBuy"] == False].copy()
        if closed_trades.empty:
            # If no closed trades, assumed no drawdown.
            mdd_results.append({"Port_IDs": port_id, "MDD": 0})
            continue
        
        # Sorted the trades in chronological order.
        closed_trades = closed_trades.sort_values(by="time")
        
        # Computed net profit for each trade: realizedProfit + fee
        # (Note: fee is negative so adding it subtracts the fee)
        closed_trades["net_profit"] = closed_trades["realizedProfit"] + closed_trades["fee"]
        
        # Computed cumulative net profit over time.
        closed_trades["cumulative_net"] = closed_trades["net_profit"].cumsum()
        
        # Compute the cumulative maximum of the cumulative net profit.
        closed_trades["cumulative_max"] = closed_trades["cumulative_net"].cummax()
        
        # Calculated drawdown at each trade.
        # If cumulative_max > 0, drawdown is (current - peak) / peak.
        # Otherwise, if no profit has been made (peak==0), we define drawdown as 0.
        closed_trades["drawdown"] = closed_trades.apply(
            lambda row: (row["cumulative_net"] - row["cumulative_max"]) / row["cumulative_max"]
            if row["cumulative_max"] > 0 else 0, axis=1)
        
        # Maximum drawdown is the minimum (most negative) drawdown observed.
        mdd = closed_trades["drawdown"].min()
        mdd_results.append({"Port_IDs": port_id, "MDD": mdd})
    
    # Created a DataFrame
    mdd_df = pd.DataFrame(mdd_results)
    return mdd_df
```

```python
def Win_Rate(new_data: pd.DataFrame) -> pd.DataFrame:
    results = []
    
    for port_id, group in new_data.groupby("Port_IDs"):
        # Prefered closed trades (activeBuy == False); if none, use all trades.
        closed_trades = group[group["activeBuy"] == False]
        trades_to_use = closed_trades if not closed_trades.empty else group
        
        total_trades = len(trades_to_use)
        # Counted winning trades: where realizedProfit > 0
        win_count = len(trades_to_use[trades_to_use["realizedProfit"] > 0])
        
        # Calculated win rate as percentage.
        rate = (win_count / total_trades) * 100 if total_trades > 0 else 0
        
        results.append({"Port_IDs": port_id, "Win Rate": rate})
    
    win_rate_df = pd.DataFrame(results)
    return win_rate_df
```

```python
def Win_Positions(new_data: pd.DataFrame) -> pd.DataFrame:
    results = []
    
    for port_id, group in new_data.groupby("Port_IDs"):
        # Used only closed trades (activeBuy == False)
        closed_trades = group[group["activeBuy"] == False]
        
        if not closed_trades.empty:
            # Counted profitable positions (realizedProfit > 0)
            wins_count = closed_trades[closed_trades["realizedProfit"].astype(float) > 0].shape[0]
            
            results.append({
                "Port_IDs": port_id,
                "Win_Positions": wins_count
            })
    
    # Created DataFrame
    win_positions_df = pd.DataFrame(results)
    return win_positions_df
```

```python
def Total_Positions(new_data: pd.DataFrame) -> pd.DataFrame:
    results = []
    
    for port_id, group in new_data.groupby("Port_IDs"):
        # Prefered closed trades (activeBuy == False) if available; otherwise, use all trades.
        closed_trades = group[group["activeBuy"] == False]
        trades_to_use = closed_trades if not closed_trades.empty else group
        
        total_trades = trades_to_use.shape[0]
        results.append({"Port_IDs": port_id, "Total Positions": total_trades})
    
    total_positions_df = pd.DataFrame(results)
    return total_positions_df
```

```python
def check(metric: str, new_data: pd.DataFrame) -> pd.DataFrame:
    # Called the appropriate metric function based on the input string.
    if metric == 'ROI':
        new_frame = ROI(new_data)
    elif metric == 'PnL':
        new_frame = PnL(new_data)
    elif metric == 'Sharpe Ratio':
        new_frame = sharpe_ratio(new_data)  
    elif metric == 'MDD':
        new_frame = MDD(new_data)
    elif metric == 'Win Rate':
        new_frame = Win_Rate(new_data)
    elif metric == 'Win Positions':
        new_frame = Win_Positions(new_data)
    elif metric == 'Total Positions':
        new_frame = Total_Positions(new_data)
    else:
        new_frame = pd.DataFrame()
    if not new_frame.empty and len(new_frame.columns) >= 2:
        metric_value_col = new_frame.columns[1]
        
    return new_frame
```

```python
metrics = ['ROI', 'PnL', 'Sharpe Ratio', 'MDD', 'Win Rate', 'Win Positions', 'Total Positions']
account_metric = None
for m in metrics:
    df_metric = check(m, new_data)
    
    # Rename the 'metrics' column to the metric name (e.g., 'ROI', 'PnL', etc.)
    df_metric = df_metric.rename(columns={"metrics": m})
    
    if account_metric is None:
        account_metric = df_metric
    else:
        account_metric = pd.merge(account_metric, df_metric, on="Port_IDs", how="outer")
```

```python
def composite_ranking(metrics_df: pd.DataFrame, weights: dict) -> pd.DataFrame:
    df = metrics_df.copy()

    for metric, weight in weights.items():
        if metric not in df.columns:
            continue  # Skip if the metric is not present
        
        # Compute z-score normalization
        mean_val = df[metric].mean()
        std_val = df[metric].std()
        if std_val == 0:
            df[metric + '_norm'] = 0
        else:
            df[metric + '_norm'] = (df[metric] - mean_val) / std_val
        
        # For MDD, lower values are better, so invert the normalized value.
        if metric == 'MDD':
            df[metric + '_norm'] = -df[metric + '_norm']
    
    # Compute the composite score as a weighted sum of normalized scores.
    composite_scores = np.zeros(len(df))
    for metric, weight in weights.items():
        if metric not in df.columns:
            continue
        composite_scores += df[metric + '_norm'] * weight
    df['Composite Score'] = composite_scores
    
    # Sort portfolios by Composite Score in descending order
    df = df.sort_values(by='Composite Score', ascending=False).reset_index(drop=True)
    
    # Create a Rank column (1 = best)
    df['Rank'] = df.index + 1
    
    # Return only the columns of interest
    return df[['Port_IDs', 'Composite Score', 'Rank']]
```
